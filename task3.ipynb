```python
import pandas as pd
df = pd.read_csv('student_feedback.csv')
df.head()
```

```python
df.isnull().sum
```

```python
df = df.drop(columns=['Rows', 'Student ID'], errors='ignore')
```

```python
df = df.dropna()
```

```python
average_scores = df.mean().sort_values(ascending=False)
print("Average Rating for Each Question:")
print(average_scores)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
average_scores.plot(kind='barh', color='lightgreen')
plt.xlabel("Average Rating (1 to 10)")
plt.title("Average Feedback for Each Evaluation Criteria")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().invert_yaxis()
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Between Feedback Parameters")
plt.show()
```

```python
top_items = average_scores.head(3)
bottom_items = average_scores.tail(3)

print("\n✅ SUMMARY REPORT")
print("Top Strengths:")
for item, val in top_items.items():
    print(f" - {item}: {val:.2f}/10")

print("\nAreas to Improve:")
for item, val in bottom_items.items():
    print(f" - {item}: {val:.2f}/10")

print("\n📌 Recommendation: Focus on improving the lowest-rated areas while maintaining excellence in the top-rated ones.")
```
